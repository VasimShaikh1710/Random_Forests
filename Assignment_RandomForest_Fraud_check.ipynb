```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/admin/Desktop/Assignment_DataScience/Fraud_check.csv')
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop = 'first', sparse = False)
df['Undergrad'] = ohe.fit_transform(df[['Undergrad']])
df['Marital.Status'] = ohe.fit_transform(df[['Marital.Status']])
df['Urban'] = ohe.fit_transform(df[['Urban']])
```

```python
df.describe()
```

```python
df.corr()
```

```python
#Add the new column as Taxinc to ----taxable_income <= 30000 as "Risky" and others are "Good"
df["TaxInc"] = pd.cut(df["Taxable.Income"], bins = [10000,30000,99620], labels = ["Risky", "Good"])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [['Risky','Good']])
df["TaxInc"] = oe.fit_transform(df[['TaxInc']])
```

```python
df.head()
```

```python
X = df.iloc[:,:6]
X
```

```python
Y = df['TaxInc']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.7)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFR = RandomForestClassifier(n_estimators =100,
                             max_features = 0.25,
                             max_samples = 0.25,
                             bootstrap = True
                             )

RFR.fit(X_train,Y_train)
Y_pred_train = RFR.predict(X_train)
Y_pred_test = RFR.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
training_accuracy = accuracy_score(Y_pred_train,Y_train)
test_accuracy = accuracy_score(Y_pred_test,Y_test)
print("training_accuracy",training_accuracy)
print("test_accuracy",test_accuracy)
```

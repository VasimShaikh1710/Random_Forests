```python
from sklearn.preprocessing import Binarizer
bn = Binarizer(threshold=7.49)
Y = bn.fit_transform(df[["Sales"]])
Y
```

```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories =[["Bad","Medium","Good"]])
df["ShelveLoc"] = oe.fit_transform(df[['ShelveLoc']])
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder(drop = 'first', sparse = False)
df["Urban"] = oe.fit_transform(df[["Urban"]])
df["US"] = oe.fit_transform(df[["US"]])
```

```python
X = df.iloc[:,1:]
X
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size = 0.7)
```

```python
from sklearn.ensemble import RandomForestRegressor
RFR = RandomForestRegressor(n_estimators =500,
                             max_features = 0.5,
                             bootstrap = False
                             )

RFR.fit(X_train,Y_train)
Y_pred_train = RFR.predict(X_train)
Y_pred_test = RFR.predict(X_test)
```

```python
R2_train = r2_score(Y_pred_train, Y_train)
R2_test = r2_score(Y_pred_test, Y_test)
print("Training Accuracy : ",R2_train)
print("Testg Accuracy : ",R2_test)
```
